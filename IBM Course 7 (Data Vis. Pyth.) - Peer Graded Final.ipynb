```python
import numpy as np
import pandas as pd

%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

print('Matplotlib version: ', mpl.__version__)
```

```python
df_DSTopics = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/Topic_Survey_Assignment.csv')
df_DSTopics
```

```python
df_DSTopics.sort_values(by='Very interested', ascending=False, axis=0, inplace=True)
df_DSTopics
```

```python
df_DSTpercent = df_DSTopics/2233*100
df_DSTpercent = round(df_DSTpercent, 2)
df_DSTpercent
```

```python
ax = df_DSTpercent.plot(kind='bar',
                    figsize=(20, 8),
                    width=0.8,
                    color=['#5cb85c', '#5bc0de', '#d9534f'],
                    fontsize=14,
                    )

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.set_title('Percentage of Respondents\' Interest in Data Sciene Areas', fontsize=16)
ax.legend(['Very Interested','Somewhat interested','Not interested'], loc='upper right', fontsize=14)

for rect in ax.patches:
        y_value = rect.get_height()
        x_value = rect.get_x() + rect.get_width() / 2
        label = (y_value)

        ax.annotate(
            '{}%'.format(label),
            (x_value, y_value+0.5),
            ha='center',
           ) 

```
